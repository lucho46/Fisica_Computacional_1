```python

# Cargar el dataset
url = "https://raw.githubusercontent.com/anferivera/Fisica_Computacional_1/refs/heads/main/Sesiones/seaborn/millas_por_galon.csv"
df = pd.read_csv(url, index_col=0)

# Mostrar información del dataset
print("=" * 60)
print("INFORMACIÓN DEL DATASET")
print("=" * 60)
print(f"Shape del dataset: {df.shape}")
print(f"\nColumnas disponibles: {df.columns.tolist()}")
print(f"\nÍndice del dataset: {df.index.name} (valores de mpg)")
print(f"\nPrimeras 5 filas:")
print(df.head())

# ==================================================================
# PREPARACIÓN DE DATOS
# ==================================================================
print("\n" + "=" * 60)
print("PREPARACIÓN DE DATOS")
print("=" * 60)

# Resetear el índice para convertir 'mpg' en una columna
df_reset = df.reset_index()
print("Dataset después de resetear el índice (mpg ahora es columna):")
print(df_reset.head())

# Ahora sí podemos acceder a mpg como columna
print(f"\nColumnas disponibles ahora: {df_reset.columns.tolist()}")

# Verificar valores nulos
print(f"\nValores nulos en mpg: {df_reset['mpg'].isnull().sum()}")
print(f"Valores nulos en horsepower: {df_reset['horsepower'].isnull().sum()}")

# Eliminar filas con valores nulos
df_clean = df_reset[['mpg', 'horsepower']].dropna()
print(f"\nFilas después de eliminar nulos: {len(df_clean)}")
print(f"Rango de mpg: {df_clean['mpg'].min():.1f} - {df_clean['mpg'].max():.1f}")
print(f"Rango de horsepower: {df_clean['horsepower'].min():.1f} - {df_clean['horsepower'].max():.1f}")

# ==================================================================
# 1. ANÁLISIS DESCRIPTIVO
# ==================================================================
print("\n" + "=" * 60)
print("1. ANÁLISIS DESCRIPTIVO")
print("=" * 60)

print("\nESTADÍSTICAS DESCRIPTIVAS:")
print(df_clean[['mpg', 'horsepower']].describe())

# ==================================================================
# 2. ANÁLISIS DE CORRELACIÓN
# ==================================================================
print("\n" + "=" * 60)
print("2. ANÁLISIS DE CORRELACIÓN")
print("=" * 60)

corr_pearson, p_valor_pearson = pearsonr(df_clean['horsepower'], df_clean['mpg'])
print(f"Coeficiente de correlación de Pearson: {corr_pearson:.4f}")
print(f"P-valor: {p_valor_pearson:.4e}")

# Interpretación de la correlación
if abs(corr_pearson) > 0.7:
    fuerza = "fuerte"
elif abs(corr_pearson) > 0.4:
    fuerza = "moderada"
else:
    fuerza = "débil"
direccion = "negativa" if corr_pearson < 0 else "positiva"
print(f"Interpretación: Correlación {fuerza} y {direccion}")

# ==================================================================
# 3. REGRESIÓN LINEAL
# ==================================================================
print("\n" + "=" * 60)
print("3. REGRESIÓN LINEAL")
print("=" * 60)

# Usando scipy.stats.linregress
slope, intercept, r_value, p_value, std_err = linregress(df_clean['horsepower'], df_clean['mpg'])

print(f"Ecuación de regresión: mpg = {intercept:.4f} + ({slope:.4f}) * horsepower")
print(f"Pendiente (β₁): {slope:.4f}")
print(f"Intersección (β₀): {intercept:.4f}")
print(f"R² (coeficiente de determinación): {r_value**2:.4f}")
print(f"Error estándar de la pendiente: {std_err:.4f}")
print(f"P-valor de la pendiente: {p_value:.4e}")

# ==================================================================
# 4. INTERVALOS DE CONFIANZA AL 95% (usando statsmodels)
# ==================================================================
print("\n" + "=" * 60)
print("4. INTERVALOS DE CONFIANZA AL 95%")
print("=" * 60)

# Preparar datos para statsmodels
X = sm.add_constant(df_clean['horsepower'])  # Añadir constante para el intercepto
y = df_clean['mpg']
model = sm.OLS(y, X).fit()

# Intervalos de confianza para los coeficientes - CORREGIDO
conf_int_coef = model.conf_int(alpha=0.05)
print("\nIntervalos de confianza al 95% para los coeficientes:")
print(f"Intercepto (β₀): [{conf_int_coef[0][0]:.4f}, {conf_int_coef[0][1]:.4f}]")
print(f"Pendiente (β₁): [{conf_int_coef[1][0]:.4f}, {conf_int_coef[1][1]:.4f}]")

# También podemos obtener los intervalos de confianza de otra forma
print("\nIntervalos de confianza (método alternativo):")
conf_int_df = model.conf_int(alpha=0.05)
conf_int_df.columns = ['2.5%', '97.5%']
conf_int_df.index = ['Intercepto', 'horsepower']
print(conf_int_df)

# Predicciones con intervalos de confianza
predicciones = model.get_prediction(X)
pred_summary = predicciones.summary_frame(alpha=0.05)

print("\nEjemplos de predicciones con IC 95% (para observaciones individuales):")
print("-" * 70)
print(f"{'Potencia':>10} {'mpg real':>10} {'mpg pred':>10} {'IC inferior':>12} {'IC superior':>12}")
print("-" * 70)

# Mostrar algunos ejemplos
ejemplos_idx = [0, 50, 100, 150, 200, 250, 300]
for idx in ejemplos_idx:
    if idx < len(df_clean):
        horse = df_clean.iloc[idx]['horsepower']
        mpg_real = df_clean.iloc[idx]['mpg']
        pred = pred_summary.iloc[idx]
        print(f"{horse:10.1f} {mpg_real:10.2f} {pred['mean']:10.2f} {pred['obs_ci_lower']:12.2f} {pred['obs_ci_upper']:12.2f}")

# ==================================================================
# 5. VISUALIZACIONES
# ==================================================================
print("\n" + "=" * 60)
print("5. GENERANDO VISUALIZACIONES...")
print("=" * 60)

# Crear figura principal
fig = plt.figure(figsize=(16, 12))
fig.suptitle('Análisis Estadístico: Millas por Galón (mpg) vs Potencia (horsepower)',
             fontsize=16, fontweight='bold')

# Grid de subplots
gs = fig.add_gridspec(2, 3, hspace=0.3, wspace=0.3)

# Subplot 1: Scatter plot con línea de regresión e IC
ax1 = fig.add_subplot(gs[0, :2])

# Puntos de datos
ax1.scatter(df_clean['horsepower'], df_clean['mpg'],
            alpha=0.5, label='Datos observados', color='steelblue', s=30)

# Crear secuencia de valores de horsepower para la línea
x_sorted = np.linspace(df_clean['horsepower'].min(), df_clean['horsepower'].max(), 100)
X_pred = sm.add_constant(x_sorted)
pred_line = model.get_prediction(X_pred).summary_frame(alpha=0.05)

# Línea de regresión
ax1.plot(x_sorted, pred_line['mean'], color='red',
         label='Línea de regresión', linewidth=2)

# Intervalo de confianza para la media
ax1.fill_between(x_sorted,
                  pred_line['mean_ci_lower'],
                  pred_line['mean_ci_upper'],
                  alpha=0.2, color='red',
                  label='IC 95% (media)')

# Intervalo de predicción
ax1.fill_between(x_sorted,
                  pred_line['obs_ci_lower'],
                  pred_line['obs_ci_upper'],
                  alpha=0.1, color='gray',
                  label='IC 95% (predicción)')

ax1.set_xlabel('Potencia (horsepower)', fontsize=12)
ax1.set_ylabel('Millas por galón (mpg)', fontsize=12)
ax1.set_title('Regresión Lineal con Intervalos de Confianza del 95%', fontsize=13)
ax1.legend(loc='upper right')
ax1.grid(True, alpha=0.3)

# Añadir ecuación y métricas
eq_text = f'mpg = {intercept:.2f} {slope:+.2f}×hp\n'
eq_text += f'R² = {r_value**2:.3f}\n'
eq_text += f'Correlación = {corr_pearson:.3f}'
ax1.text(0.05, 0.95, eq_text, transform=ax1.transAxes, fontsize=11,
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# Subplot 2: Distribución de variables
ax2 = fig.add_subplot(gs[0, 2])
ax2.hist(df_clean['mpg'], bins=20, alpha=0.7, color='steelblue', edgecolor='black')
ax2.set_xlabel('mpg')
ax2.set_ylabel('Frecuencia')
ax2.set_title('Distribución de mpg')
ax2.grid(True, alpha=0.3)

# Subplot 3: Histograma de residuos
ax3 = fig.add_subplot(gs[1, 0])
residuos = model.resid
ax3.hist(residuos, bins=25, alpha=0.7, color='coral', edgecolor='black')
ax3.axvline(x=0, color='black', linestyle='--', linewidth=1)
ax3.set_xlabel('Residuos')
ax3.set_ylabel('Frecuencia')
ax3.set_title('Distribución de Residuos')

# Prueba de normalidad
shapiro_stat, shapiro_p = stats.shapiro(residuos)
norm_text = f'Shapiro-Wilk p-valor: {shapiro_p:.4f}'
norm_text += '\n' + ('Normal' if shapiro_p > 0.05 else 'No normal')
ax3.text(0.05, 0.95, norm_text, transform=ax3.transAxes, fontsize=9,
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# Subplot 4: Q-Q plot
ax4 = fig.add_subplot(gs[1, 1])
stats.probplot(residuos, dist="norm", plot=ax4)
ax4.set_title('Gráfico Q-Q de Residuos')
ax4.grid(True, alpha=0.3)

# Subplot 5: Residuos vs valores ajustados
ax5 = fig.add_subplot(gs[1, 2])
fitted = model.fittedvalues
ax5.scatter(fitted, residuos, alpha=0.5, color='purple', s=20)
ax5.axhline(y=0, color='black', linestyle='--', linewidth=1)
ax5.set_xlabel('Valores ajustados')
ax5.set_ylabel('Residuos')
ax5.set_title('Residuos vs Valores Ajustados')
ax5.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('analisis_completo_mpg_horsepower.png', dpi=300, bbox_inches='tight')
plt.show()

# ==================================================================
# 6. GRÁFICO ADICIONAL: IC para puntos específicos
# ==================================================================
fig2, ax = plt.subplots(figsize=(12, 8))

# Seleccionar puntos para mostrar IC
np.random.seed(42)
indices = np.random.choice(len(df_clean), size=15, replace=False)
df_sample = df_clean.iloc[indices].sort_values('horsepower')

# Predicciones para la muestra
X_sample = sm.add_constant(df_sample['horsepower'])
pred_sample = model.get_prediction(X_sample).summary_frame(alpha=0.05)

# Gráfico
ax.errorbar(df_sample['horsepower'], pred_sample['mean'],
            yerr=[pred_sample['mean'] - pred_sample['obs_ci_lower'],
                  pred_sample['obs_ci_upper'] - pred_sample['mean']],
            fmt='o', color='red', capsize=5, capthick=2,
            label='Predicción con IC 95%', markersize=8)
ax.scatter(df_sample['horsepower'], df_sample['mpg'],
          color='blue', alpha=0.7, s=100, label='Valores reales', zorder=5)

# Línea de regresión
x_line = np.linspace(df_clean['horsepower'].min(), df_clean['horsepower'].max(), 100)
X_line = sm.add_constant(x_line)
pred_line = model.get_prediction(X_line).summary_frame(alpha=0.05)
ax.plot(x_line, pred_line['mean'], 'k-', linewidth=2, label='Línea de regresión')
ax.fill_between(x_line, pred_line['obs_ci_lower'], pred_line['obs_ci_upper'],
                alpha=0.2, color='gray', label='IC 95% (predicción)')

ax.set_xlabel('Potencia (horsepower)', fontsize=14)
ax.set_ylabel('Millas por galón (mpg)', fontsize=14)
ax.set_title('Predicciones con Intervalos de Confianza del 95% (15 puntos aleatorios)', fontsize=14)
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('ic_puntos_especificos.png', dpi=300, bbox_inches='tight')
plt.show()

# ==================================================================
# 7. RESUMEN ESTADÍSTICO
# ==================================================================
print("\n" + "=" * 60)
print("6. RESUMEN ESTADÍSTICO COMPLETO")
print("=" * 60)
print(model.summary().tables[0])
print("\n" + str(model.summary().tables[1]))

# ==================================================================
# 8. CONCLUSIONES
# ==================================================================
print("\n" + "=" * 60)
print("CONCLUSIONES")
print("=" * 60)

print(f"""
ANÁLISIS DE LA RELACIÓN ENTRE CONSUMO (mpg) Y POTENCIA (horsepower)
{'='*60}

1. RELACIÓN ENTRE VARIABLES:
   - Coeficiente de correlación: r = {corr_pearson:.3f}
   - Dirección: {'Negativa' if corr_pearson < 0 else 'Positiva'}
     (a mayor potencia, {'menor' if corr_pearson < 0 else 'mayor'} consumo)
   - Fuerza: {fuerza.upper()}
   - Significancia estadística: {'SÍ' if p_valor_pearson < 0.05 else 'NO'} (p-valor = {p_valor_pearson:.4e})

2. MODELO DE REGRESIÓN LINEAL:
   - Ecuación: mpg = {intercept:.2f} {slope:+.2f} × horsepower
   - R² = {r_value**2:.3f} (el modelo explica el {r_value**2*100:.1f}% de la variabilidad)
   - Error estándar de la estimación: {np.sqrt(model.mse_resid):.2f} mpg

3. INTERPRETACIÓN DE COEFICIENTES CON IC 95%:
   - Pendiente (β₁): Por cada aumento de 1 unidad en horsepower,
     el consumo disminuye en promedio {abs(slope):.3f} millas por galón.
   - Intervalo de confianza 95% para β₁: [{conf_int_coef[1][0]:.4f}, {conf_int_coef[1][1]:.4f}]
   - Intervalo de confianza 95% para β₀: [{conf_int_coef[0][0]:.4f}, {conf_int_coef[0][1]:.4f}]

4. VALIDACIÓN DEL MODELO:
   - Normalidad de residuos: {'SÍ' if shapiro_p > 0.05 else 'NO'} se cumple (p-valor = {shapiro_p:.4f})
   - Homocedasticidad: Verificar gráfico de residuos vs ajustados
   - Independencia: Asumida por el diseño del estudio

5. LIMITACIONES:
   - La relación podría no ser lineal en todo el rango
   - Existen otros factores que afectan el consumo (peso, cilindros, año)
   - Valores atípicos pueden influir en los resultados
""")

print("\n" + "=" * 60)
print("ANÁLISIS COMPLETADO")
print("=" * 60)
```
