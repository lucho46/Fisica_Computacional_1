```python
import seaborn as sns

data = sns.load_dataset("penguins").dropna().values
df = pd.DataFrame(data, columns=["especie", "isla", "longitud_pico (mm)", "profundidad_pico (mm)", "longitud_aleta (mm)", "masa_corporal (g)", "genero"])

df
```

¿Cuáles son las profundidades del pico máxima y mínima de los pinguinos de la isla Torgersen?

```python
def maxim(list):
  maximo = list[0]
  for i in range(len(list)):
    if maximo < list[i]:
      maximo = list[i]
  return maximo

def mini(list):
  mini = list[0]
  for i in range(len(list)):
    if mini > list[i]:
      mini = list[i]
  return mini

print(f"la profundidad de pico maxima es : {maxim(df["profundidad_pico (mm)"])} \nla profundidad de pido minima es :  {mini(df["profundidad_pico (mm)"])}")
```

¿De qué isla, especie y género es el pingüino con la mayor masa corporal? Ayuda: utilice la función argmax()

```python
index = np.argmax(df["masa_corporal (g)"])
print(f"Pinguino con mayor peso corporal: \n isla: {df["isla"][index]} \n especie: {df["especie"][index]} \n genero: {df["genero"][index]} ")
```

¿Cuál es la masa corporal media de los pinguinos macho que no son de la especie Gentoo?

```python
listax = []
for i in range(len(df)):
  if df["especie"][i] == "Gentoo" or df["genero"][i] == "Female":
    listax.append(i)

df1 = df.drop(df.index[listax])

media = np.mean(df1["masa_corporal (g)"])
print(f"la masa media de los pinguinos macho que no son de la especie Gentoo es : \n {media} g")
```

Grafique la distribución de la longitud del pico. Añada el valor medio y la desviación estándar en el título de la figura. Ayuda: utilice la funcion hist()

```python
des = np.std(df["longitud_pico (mm)"])
media = np.mean(df["longitud_pico (mm)"])
x = np.zeros(2) + media
y = np.array([0,25])

plt.hist(df["longitud_pico (mm)"], 30 , rwidth = 0.8, alpha = 0.6)
plt.title(f"Longitud del pico  $ \mu$ = {media:.2f} a = {des:.2f}")
plt.legend()
plt.show()
```

¿Cuáles son las islas y especies presentes en el conjunto de datos? ¿Con qué frecuencia aparecen? Grafique la distribución de la masa corporal para los pingüinos.

```python
def datos(lista):
  xlista = []
  for i in range(len(lista)):
    if lista[i] not in xlista:
      xlista.append(lista[i])

  xfrec = []
  for i in range(len(xlista)):
    x = 0
    for j in range(len(lista)):
      if lista[j] == xlista[i]:
        x += 1
    xfrec.append(x)

  return xlista , xfrec

print(f"las islas y especies del conjunto de datos es: \n islas: {datos(df["isla"])[0]} con frecuencias: {datos(df["isla"])[1]} respectivamente \n especies: {datos(df["especie"])[0]} con frecuencias: {datos(df["especie"])[1]} respectivamente")
```

```python
norm = np.mean(df["masa_corporal (g)"])
std = np.std(df["masa_corporal (g)"])

plt.hist(df["masa_corporal (g)"], 30 , rwidth = 0.8, alpha = 0.6)
plt.title(f"Masa corporal (g)  $ \mu$ = {norm:.2f} a = {std:.2f}")
plt.legend()
plt.show()
```
