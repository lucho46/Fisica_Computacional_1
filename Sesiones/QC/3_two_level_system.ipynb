# Two level system in QM

In `quantum mechanics`, a **two-state system** (also known as a two-level system) is a quantum system that can exist in any quantum superposition of two independent (physically distinguishable) quantum states. The Hilbert space describing such a system is two-dimensional. Any two-state system can also be seen as a `qubit`.

* A well known example of a two-state system is the **spin** of a spin-1/2 particle such as an electron, whose spin can have values +ħ/2 or −ħ/2, where ħ is the reduced Planck constant. The **spin** of the electron in which the two levels can be taken as spin up and spin down.
* the **polarization** of a single photon in which the two spin states (left-handed and the right-handed circular polarization).
* **NH3** system
* ...

Supposing the two available basis states of the system are $(\ket{0}\,,\ket{1})$ or $(\ket{1}\,,\ket{2})$ with a `state` or `vector` $\ket{\psi}$ written as a superposition of these two states

$$\ket{\psi}=c_1\ket{1}+c_2\ket{2}$$

These two complex numbers $c_i$ may be considered coordinates in a two-dimensional complex `Hilbert space`

$$
\begin{align}
\ket{\psi}
=
\begin{pmatrix}
c_1 \\
c_2
\end{pmatrix}= & c_1
\begin{pmatrix}
1 \\
0
\end{pmatrix} + c_2
\begin{pmatrix}
0 \\
1
\end{pmatrix}\\
= & \, c_1\ket{1}\,+\,c_2\ket{2}
\end{align}
$$

The norm of the wave function $\psi$ is one, i.e. $\bra{\psi}\ket{\psi} = \sum_i|c_i|^2 = |c_1|^2+|c_2|^2=1$.

### Energy

In the case of energy and the corresponding Hamiltonian $H$ (ermitian operator)

$$
H_{ij}=\bra{i}H\ket{j} = 
\begin{pmatrix}
\bra{1}H\ket{1} & \bra{1}H\ket{2}\\
\bra{2}H\ket{1} & \bra{2}H\ket{2}
\end{pmatrix} =
\begin{pmatrix}
H_{11} & H_{12}\\
H_{12}^* & H_{22}
\end{pmatrix}
$$

and the time-independent Schrödinger equation states $H\ket{\psi}=E\ket{\psi}$ is a system of two linear equations:
 $$
\begin{pmatrix}
H_{11} & H_{12}\\
H_{12}^* & H_{22}
\end{pmatrix}\begin{pmatrix}
c_1\\
c_2
\end{pmatrix} = E \begin{pmatrix}
c_1\\
c_2
\end{pmatrix}
$$

### Eigenvalues of H

The most general form of a $2 \times 2$ Hermitian matrix such as the Hamiltonian of a two-state system is given by
$$
H =
\begin{pmatrix}
H_{11} & c_x - ic_y \\
c_x + ic_y & H_{22}
\end{pmatrix}
$$
where $H_{11},H_{22}, c_x$ and $c_y$ are `real numbers` with units of `energy` such that $c_x+ic_y=H_{12}$. 

The allowed energy levels of the system, namely the eigenvalues of the Hamiltonian matrix, can be found as:

This matrix can be decomposed as
$$
\begin{align}
H =& \alpha \cdot \sigma_0 + c_x \cdot \sigma_1 + c_y \cdot \sigma_2 + c_z \cdot \sigma_3
\\
=& 
\alpha\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
+c_x\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
+c_y\begin{pmatrix}
0 & -i \\
i & 0
\end{pmatrix}
+c_z\begin{pmatrix}
1 & 0 \\
0 & -1
\end{pmatrix},\\
=&\begin{pmatrix}
\alpha + c_z & c_x - ic_y \\
c_x + ic_y & \alpha - c_z
\end{pmatrix}.
\end{align}
$$

Here,
$$
\alpha = \frac{1}{2}(H_{11} + H_{22}), 
\quad \text{and} \quad
c_z = \frac{1}{2}(H_{11} - H_{22})
$$
are real numbers. The matrix $\sigma_0$ is the 2×2 identity matrix and the matrices $\sigma_k$ are the `Pauli matrices`. This decomposition simplifies the analysis of the system, especially in the time-independent case.

The Hamiltonian can be further condensed as
$$
H = \alpha \cdot \sigma_0 + \vec{\mathbf{c}} \cdot \vec{\boldsymbol{\sigma}}\,
$$

The vecto $\vec{c}=(c_x,c_y,c_z)$. This representation simplifies the analysis of the time evolution of the system.

The corresponding characteristic equation for the system's energy level is

$$
\begin{vmatrix}
\alpha + c_z -E & c_x - ic_y \\
c_x + ic_y & \alpha - c_z -E
\end{vmatrix} = 0 
$$
$$
(\alpha + c_z -E)(\alpha - c_z -E)-(c_x^2+c_y^2)=0\\
(\alpha-E)^2 - c_z^2-(c_x^2+c_y^2)= (\alpha-E)^2 - c_i^2 =0
$$


If the two-state system's time-independent Hamiltonian $H$ is defined as above, then its `eigenvalues` are given by
$$
\boxed{
E_{\pm} = \alpha \pm |\mathbf{c}|}\\
 =  \frac{1}{2}(H_{11} + H_{22}) \pm (H_{12}^2+\dfrac{1}{4}(H_{11}-H_{22})^2)^{1/2} = E_0\pm \Delta E/2
$$

The corresponding `eigenvectors` are denoted as $|+\rangle$ and $|-\rangle$.

**Level spliting**

$$
\boxed{
\Delta E = E_+-E_- = 2|c|=\sqrt{(H_{11}-H_{22})^2+4\,H_{12}^2}}
$$

## Time dependent. 

We now assume that the probability amplitudes are time-dependent, though the basis states are not. The Time-dependent Schrödinger equation states $i\hbar\partial_t\ket{\psi}=H\ket{\psi}$. Using the evolution temporal operator

$$
c(t) = e^{-iH t / \hbar}c(0) = U(t)c_0
$$

where $U(t)$ is the evolution matrix. It can be shown that

$$
U(t)=
c(t) = e^{-iH t / \hbar} = e^{-i(\alpha\cdot\sigma_0 +\vec{c}\cdot\vec{\sigma}) t / \hbar}
= e^{-i\alpha t / \hbar} \bigg(\cos(\dfrac{|\vec{c}|t}{\hbar})\sigma_0-i\sin(\dfrac{|\vec{c}|t}{\hbar})\hat{r}\cdot\sigma\bigg).
$$

The last equation is general, however we could change to the basis of the eigenvectors of the Hamiltonian. In other words, where $\ket{1}$ and $\ket{2}$ are chosen to be the eigenvectors, therefore

$$
H = \begin{pmatrix}
E_1 & 0 \\
0 & E_2
\end{pmatrix}
$$

$$
U(t)=
c(t) = e^{-iH t / \hbar} 
= e^{-i\alpha t / \hbar} \bigg(\cos(\dfrac{|c|t}{\hbar})\sigma_0-i\sin(\dfrac{|c|t}{\hbar})\dfrac{c_z}{|c|}\sigma_3\bigg)
$$

The factor $e^{-iH t / \hbar} $ factor merely contributes to the overall phase of the operator, and can usually be ignored to yield a new time evolution operator that is physically indistinguishable from the original operator.

## Quantum oscillation. 

**Rabi formula for static perturbations** 

Suppose that the system starts in one of the basis states at $t=0$, it is $\ket{1}=c(0)=\begin{pmatrix}1\\0\end{pmatrix}$ and we are interested in the probability of occupation of each of the basis states as a function of time when $H$ is the time-independent Hamiltonian.

$$
c(t)=U(t)c(0) = 
\begin{pmatrix}
U_{11}(t) & U_{12}(t)\\
U_{21}(t) & U_{22}(t)
\end{pmatrix}
\begin{pmatrix} 1 \\ 0
 \end{pmatrix} =
 \begin{pmatrix}
U_{11}(t) \\
U_{21}(t)
\end{pmatrix}
$$

The **probability of occupation** of state $i$ is $P_i(t)=|c_i(t)|^2=|U_{i1}(t)|^2$

\begin{align}
U_{11}(t) =& \cos(\dfrac{|c|t}{\hbar}) -i\sin(\dfrac{|c|t}{\hbar})\dfrac{c_z}{|c|}\\
c_x= & \dfrac{(H_{12}-H_{21})}{2}\,, c_y=i\dfrac{(H_{12}-H_{21})}{2}\,, c_z = & \dfrac{(H_{11}-H_{22})}{2}
\end{align}

Where
\begin{align}
\Omega =& \dfrac{c}{\hbar} =  \sqrt{(c_x^2+c_y^2+c_z^2)/\hbar^2} = \sqrt{|(c_x+ic_y)/\hbar|^2+(c_z/\hbar)}= \sqrt{|\Omega_R|^2+\Delta^2}\\
\Omega_R =& \dfrac{c_x+ic_y}{\hbar}\,\, \text{ is called the Rabi frequency,}\\
\Delta =& \dfrac{c_z}{\hbar}\,\, \text{ is called the detuning}\,.
\end{align}

\begin{align}
P_1(t) = |U_{11}(t)|^2 = \cos^2(\Omega t)+\sin^2(\Omega t)\dfrac{\Delta^2}{\Omega^2}
\end{align}

At zero detuning:
$$
P_1(t) \approx \cos^2(|\Omega_R| t)
$$

i.e., there is Rabi flopping from guaranteed occupation of state 1, to guaranteed occupation of state 2, and back to state 1, etc., with frequency $\Omega_R$. As the detuning is increased away from zero, the frequency of the flopping increases to $(\Omega)$ and the amplitude of exciting the electron decreases to $\Omega^2/\Delta^2$. For time dependent Hamiltonians induced by light waves, see the articles on `Rabi cycle`.

### Example:

```python
import numpy as np
import matplotlib.pyplot as plt

# Example of detuning effects on Rabi oscillations:
# Benchmark parameters
Omega0 = 0.5
Delta0 = 0
Delta1 = Omega0/2
Delta2 = Omega0
t = np.linspace(0, 4*np.pi, 100)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(t, np.cos(Omega0*t)**2+(np.sin(Omega0*t)*Delta0/Omega0)**2, label='$\\Delta=0, \\to \\Omega_R=\\Omega=1/2$')
plt.plot(t, np.cos(Omega0*t)**2+(np.sin(Omega0*t)*Delta1/Omega0)**2, label='$\\Delta=\\Omega/2, \\to \\Omega_R^2=3/4\\Omega^2$')
plt.plot(t, np.cos(Omega0*t)**2+(np.sin(Omega0*t)*Delta2/Omega0)**2, label='$\\Delta=\\Omega, \\to \\Omega_R=0$ no physics')
plt.xlabel('t')
plt.ylabel('$P_1(t)$')
plt.title('Ω_Rabi as a function of time for different detunings')
plt.xlim(t.min(), t.max())
plt.legend()
####################################################################
plt.subplot(1, 2, 2)
plt.plot(t, 1- (np.cos(Omega0*t)**2+(np.sin(Omega0*t)*Delta0/Omega0)**2), label='$\\Delta=0, \\to \\Omega_R=\\Omega=1/2$')
plt.plot(t, 1- (np.cos(Omega0*t)**2+(np.sin(Omega0*t)*Delta1/Omega0)**2), label='$\\Delta=\\Omega/2, \\to \\Omega_R^2=3/4\\Omega^2$')
plt.plot(t, 1- (np.cos(Omega0*t)**2+(np.sin(Omega0*t)*Delta2/Omega0)**2), label='$\\Delta=\\Omega, \\to \\Omega_R=0$')
plt.xlabel('t')
plt.ylabel('$P_2(t)=1-P_1(t)$')
plt.title('Probability of being in state $|2\\rangle$')
plt.xlim(t.min(), t.max())
plt.legend()
plt.show()  
```

### Rabi rutine

```python
hbar = 1.0545718e-34 # Planck's constant over 2 pi in J*s
#Rabi frequency:
def Omega_Rabi(cx,cy):
    return np.sqrt(cx**2 + cy**2)/hbar
#Detuning:
def Delta(cz):
    return np.abs(cz)/hbar
#Generalized Rabi frequency:
def Omega(cx,cy,cz):
    return np.sqrt(Omega_Rabi(cx,cy)**2 + Delta(cz)**2)
#Probability of being in state |1> at time t:
def P1(cx,cy,cz,t):
    return np.cos(Omega(cx,cy,cz)*t)**2 + np.sin(Omega(cx,cy,cz)*t)**2*(Delta(cz)/Omega(cx,cy,cz))**2
```

```python
plt.figure(figsize=(12, 4))
###################### First case without detuning ###############
plt.subplot(1, 2, 1)
# Benchmark parameters
omega0 = 0.5
cx = omega0*hbar/np.sqrt(2)
cy = cx
cz = 0
print(f"Rabi frequency: {Omega_Rabi(cx,cy):.3e} Hz")
print(f"Detuning: {Delta(cz):.3e} Hz")
print(f"Generalized Rabi frequency: {Omega(cx,cy,cz):.3e} Hz\n")

t = np.linspace(0, 4*np.pi, 100)
Pr = P1(cx,cy,cz,t)
plt.plot(t, Pr, label='P1 with cx=1, cy=0, cz=0')
plt.xlabel('t')
plt.ylabel('$P_1(t)$')
plt.xlim(t.min(), t.max())
plt.ylim(0, 1)
plt.title(f'Rabi Oscillations: $\\Omega_R={Omega_Rabi(cx,cy):.3e}$ Hz, $\\Delta={Delta(cz):.3e}$ Hz')

###################### Second case with detuning ###############
plt.subplot(1, 2, 2)
# Benchmark parameters
omega0 = 0.5
cx = np.sqrt(3/4)*omega0*hbar/np.sqrt(2)
cy = cx
#cz = np.sqrt(omega0*2-Omega_Rabi(cx,cy)**2)*hbar
cz = omega0*hbar/2

print(f"Rabi frequency: {Omega_Rabi(cx,cy):.3e} Hz")
print(f"Detuning: {Delta(cz):.3e} Hz")
print(f"Generalized Rabi frequency: {Omega(cx,cy,cz):.3e} Hz")

t = np.linspace(0, 4*np.pi, 100)
Pr = P1(cx,cy,cz,t)
plt.plot(t, Pr, color='orange', label='P1 with cx=1, cy=0, cz=0')
plt.xlabel('t')
plt.ylabel('$P_1(t)$')
plt.xlim(t.min(), t.max())
plt.ylim(0, 1)
plt.title(f'Rabi Oscillations: $\\Omega_R={Omega_Rabi(cx,cy):.3e}$ Hz, $\\Delta={Delta(cz):.3e}$ Hz')
plt.show()
```

## State flipping (spin flipping)

[see wiki:](https://en.wikipedia.org/wiki/Rabi_cycle)
One example of Rabi flopping is the spin flipping within a quantum system containing a spin-1/2 particle and an oscillating magnetic field $\vec{B}= B_0 \hat{k}+B_1(\cos(\omega t)\hat{i}+\sin(\omega t)\hat{j})$. In this case the Hamiltonian is 

$$
H = -\vec{\mu}\cdot\vec{B} = \dfrac{\hbar}{2}
\begin{pmatrix}
\omega_0 & \omega_1 e^{-i\omega t} \\
\omega_1 e^{+i\omega t} & -\omega_0
\end{pmatrix}
$$

where $\omega_0 = \gamma B_0$ and $\omega_1 = \gamma B_1$ with $\gamma$ is the gyromagnetic ratio of the particle.

In this case. The probability that a particle starting in the spin up state flips to the spin down state can be stated as

$$
P(t) = \dfrac{\omega_1^2}{\Omega^2}\sin^2\big(\dfrac{\Omega t}{2}\big)
$$

where $\Omega=\sqrt{\Delta \omega^2 + \omega_1^2}=\sqrt{(\omega -\omega_0)^2 + \omega_1^2}$
