# Quantum logic gates

1. In a clasical computer a Logic gates perform basic logical operations and serve as the building blocks of `digital circuits`. Most logic gates take one or two binary values `(0 or 1)` as inputs and produce a single binary output `bits`. Basic circuits contain only a few logic gates, while complex ones — such as microprocessors — may contain millions (see [1_logic_gates.ipynb](1_logic_gates.ipynb )).

2. In quantum computing (QC) a quantum logic gate (or simply quantum gate) is a basic quantum circuit operating on a small number of `qubits`. **Quantum logic gates are the building blocks of quantum circuits**, like classical logic gates are for conventional digital circuits.
A general wave function (the *qubit*) is a coherent superposition 

$$\ket{\psi}=c_0\ket{0}+c_1\ket{1}$$
$$
\ket{0} =\begin{pmatrix}1\\0\end{pmatrix}\,,\hspace{0.5 cm}
\ket{1} =\begin{pmatrix}0\\1\end{pmatrix}
$$

Quantum gates are `unitary operators` on a Hilbert space, and are described as unitary matrices relative to some orthonormal basis. Usually the computational basis is: $\ket{0}\, \ket{1}\, \ket{d-1}\cdots$.

[Common quantum logic gates, circuit form(s) and the corresponding unitary matrices](https://upload.wikimedia.org/wikipedia/commons/e/e0/Quantum_Logic_Gates.png)

<center><img src="https://upload.wikimedia.org/wikipedia/commons/e/e0/Quantum_Logic_Gates.png" width="600"><\center>

*  A gate that acts on 
n qubits (a register) is represented by a 
$2^{n}\times 2^{n}$ unitary matrix.
* The basis vectors (eigenstates) are the possible outcomes if the state of the qubits is measured, and a quantum state is a linear combination of these outcomes.
* The **action of the gate** on a specific quantum state is found by multiplying the vector 
$\ket{\psi_1}$, which represents the state by the matrix 
U representing the gate. The result is a new quantum state $\ket{\psi_2}$

$$U\ket{\psi_1}=\ket{\psi_2}$$



## 1. Pauli gates (X,Y,Z)

They referes to `Pauli matrices` $\sigma_i$ and act on a single qubit. They represent a rotation around the $x$, $y$ and $z$ axes of the Bloch sphere by $\pi$ radians.

**Pauli-X**

\begin{align}
X\,\ket{0}=&
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}\begin{pmatrix}
1\\
0
\end{pmatrix}=\begin{pmatrix}
0\\
1
\end{pmatrix}=\ket{1} \to X\,\ket{0}=\ket{1} \\
X\,\ket{1}=&
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}\begin{pmatrix}
0\\
1
\end{pmatrix}=\begin{pmatrix}
1\\
0
\end{pmatrix}=\ket{0} \to X\,\ket{1}=\ket{0}
\end{align}

NOTA:
1. The Pauli-X gate is the quantum equivalent of the `NOT gate` for classical computers
2. Rotation around the z axis on the Bloch sphere.
3. It is a bit-flip as it maps $\ket{0}\to\ket{1}$ and $\ket{1}\to\ket{0}$. Similarly

**Pauli-Y**

\begin{align}
Y\,\ket{0}=&
\begin{pmatrix}
0 & -i \\
i & 0
\end{pmatrix}\begin{pmatrix}
1\\
0
\end{pmatrix}=i\begin{pmatrix}
0\\
1
\end{pmatrix}=i\ket{1} \to Y\,\ket{0}=i\ket{1} \\
Y\,\ket{1}=&
\begin{pmatrix}
0 & -i \\
i & 0
\end{pmatrix}\begin{pmatrix}
0\\
1
\end{pmatrix}=-i\begin{pmatrix}
1\\
0
\end{pmatrix}=\ket{0} \to Y\,\ket{1}=-i\ket{0}
\end{align}

**Pauli-Z**

\begin{align}
Z\,\ket{0}=&
\begin{pmatrix}
1 & 0 \\
0 & -1
\end{pmatrix}\begin{pmatrix}
1\\
0
\end{pmatrix}=+\begin{pmatrix}
1\\
0
\end{pmatrix}=+\ket{0} \to Z\,\ket{0}=+\ket{0} \\
Z\,\ket{1}=&
\begin{pmatrix}
1 & 0 \\
0 & -1
\end{pmatrix}\begin{pmatrix}
0\\
1
\end{pmatrix}=-\begin{pmatrix}
0\\
1
\end{pmatrix}=-\ket{1} \to Z\,\ket{1}=-\ket{1}
\end{align}

NOTA: Pauli Z is sometimes called `phase-flip`.

**Hadamard gate**

the Hadamard gate is a one-qubit rotation, mapping the qubit-basis states 
$\ket{0}$ and $\ket{1}$ to two superposition states with equal weight of the computational basis states 
$\ket{0}$ and $\ket{1}$. acts on a single qubit. It maps the basis states

\begin{align}
\ket{0} \to \ket{+}=\dfrac{\ket{0}+\ket{1}}{\sqrt{2}}\, \ket{1} \to \ket{-}=\dfrac{\ket{0}-\ket{1}}{\sqrt{2}}
\end{align}

\begin{align}
H\,\ket{0}=&\dfrac{1}{\sqrt{2}}
\begin{pmatrix}
1 & 1 \\
1 & -1
\end{pmatrix}\begin{pmatrix}
1\\
0
\end{pmatrix}=+\dfrac{1}{\sqrt{2}}\left[\begin{pmatrix}
1\\
0
\end{pmatrix}+
\begin{pmatrix}
0\\
1
\end{pmatrix}\right]=\dfrac{\ket{0}+\ket{1}}{\sqrt{2}}
=\ket{+} \to H\,\ket{0}= \ket{+} 
\end{align}


\begin{align}
H\,\ket{1}=&\dfrac{1}{\sqrt{2}}
\begin{pmatrix}
1 & 1 \\
1 & -1
\end{pmatrix}\begin{pmatrix}
0\\
1
\end{pmatrix}=+\dfrac{1}{\sqrt{2}}\left[\begin{pmatrix}
1\\
0
\end{pmatrix}-
\begin{pmatrix}
0\\
1
\end{pmatrix}\right]=\dfrac{\ket{0}-\ket{1}}{\sqrt{2}}
=\ket{-} \to H\,\ket{1}= \ket{-} 
\end{align}

NOTA: 

*  (it creates an equal superposition state if given a computational basis state).
* The Hadamard gate performs a rotation of $\pi$ about the axis $(\hat{x}+\hat{y})/\sqrt{2}$ at the Bloch sphere
* Hadamard gate is used to perform a change of basis. For example, 
$HZH = X$ and $H\sqrt{X}H=\sqrt{Z}=S$
