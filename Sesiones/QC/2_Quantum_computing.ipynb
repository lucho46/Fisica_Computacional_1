A `quantum computer` is a real or theoretical computer that exploits quantum phenomena like superposition and entanglement in an essential way. 
* It is widely believed that a quantum computer could perform some calculations exponentially faster than any classical computer. For example, a large-scale quantum computer could break some widely used encryption schemes and aid physicists in performing physical simulations. 
* Current hardware implementations of quantum computation are largely experimental and only suitable for specialized tasks.
* The basic unit of information in quantum computing, the `qubit` (or **quantum bit**), serves the same function as the bit in ordinary or "classical" computing.
* Quantum computers are not yet practical for real-world applications
* If a physical qubit is not sufficiently isolated from its environment, it suffers from quantum decoherence, introducing noise into calculations.

## States of qubit

Whereas the state of a bit can only be binary (either 0 or 1), the general state of a qubit according to `quantum mechanics` (QM) can be $\ket{0}$ or $\ket{1}$. 

The state of a system can be a
* **Coherent superposition** ... linear combinations of the of solution of the Scródinger equation
* *it can be in both states simultaneously*. ?

A `qubit is a two-state (or two-level) quantum-mechanical system`.
It can be represented by a linear superposition of its `two orthonormal basis states`. These vectors are usually denoted as:
$$
\ket{0} \to \text{ket 0}  \\
\ket{1} \to \text{ket 1} 
$$

The elements of the dual space are:

$$
\bra{0} \to \text{bra 0}  \\
\bra{1} \to \text{bra 1} 
$$

and the orthogonal relation of this two states is $\bra{i}\ket{j}=\delta_{ij}$.

Diagramatically those states are:
[quibits](https://upload.wikimedia.org/wikipedia/commons/thumb/4/4e/Simple_qubits.svg/1920px-Simple_qubits.svg.png)
<center><img src="https://upload.wikimedia.org/wikipedia/commons/thumb/4/4e/Simple_qubits.svg/1920px-Simple_qubits.svg.png" width="200"></center>

A general wave function (the *qubit*) is a coherent superposition $\ket{\psi}=c_0\ket{0}+c_1\ket{1}$
where $c_i=\bra{i}\ket{\psi}$ are two complex numbers $c_i$ (probability amplitudes) considered as coordinates in a two-dimensional complex `Hilbert space` with the next `matrix basis`
$$
\ket{0} =\begin{pmatrix}1\\0\end{pmatrix}\,,\hspace{0.5 cm}
\ket{1} =\begin{pmatrix}0\\1\end{pmatrix}
$$
In this basis, the wave function is:
$$
\begin{align}
\ket{\psi}
=
\begin{pmatrix}
c_0 \\
c_1
\end{pmatrix}= & c_0
\begin{pmatrix}
1 \\
0
\end{pmatrix} + c_1
\begin{pmatrix}
0 \\
1
\end{pmatrix}\\
= & \, c_0\ket{0}\,+\,c_1\ket{1}
\end{align}
$$


If the state $\ket{\psi}$ is normalized, the norm of the state vector is unity, i.e. $\bra{\psi}\ket{\psi} = \sum_i|c_i|^2 = |c_0|^2+|c_1|^2=1$.

Qubit basis states can also be combined to form product basis states. For example, two qubits could be represented in a four-dimensional linear vector space spanned by the following product basis states:

$$
\ket{00} =\begin{pmatrix}1\\0\\0\\0\end{pmatrix}\,,\hspace{0.2 cm}
\ket{01} =\begin{pmatrix}0\\1\\0\\0\end{pmatrix}\,,\hspace{0.2 cm}
\ket{10} =\begin{pmatrix}0\\0\\1\\0\end{pmatrix}\,,\hspace{0.2 cm}
\ket{11} =\begin{pmatrix}0\\0\\0\\1\end{pmatrix}\,,\hspace{0.2 cm}
$$
In general, $n$ qubits are represented by a superposition state vector in $2^n$ dimensional Hilbert space.

## Two level system in QM

In `quantum mechanics`, a **two-state system** (also known as a two-level system) is a quantum system that can exist in any quantum superposition of two independent (physically distinguishable) quantum states. The Hilbert space describing such a system is two-dimensional. Any two-state system can also be seen as a `qubit`.

* A well known example of a two-state system is the **spin** of a spin-1/2 particle such as an electron, whose spin can have values +ħ/2 or −ħ/2, where ħ is the reduced Planck constant. The **spin** of the electron in which the two levels can be taken as spin up and spin down.
* the **polarization** of a single photon in which the two spin states (left-handed and the right-handed circular polarization).
* **NH3** system
* ...

Supposing the two available basis states of the system are $(\ket{0}\,,\ket{1})$ or $(\ket{1}\,,\ket{2})$ with a `state` or `vector` $\ket{\psi}$ written as a superposition of these two states

$$\ket{\psi}=c_1\ket{1}+c_2\ket{2}$$
