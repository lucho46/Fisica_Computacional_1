<h1> Basic Introduction to c++</h1>

C++ is a compiled, general-purpose programming language designed as an extension of the C language with support for object-oriented programming.It is known for its high performance, speed, and detailed control of hardware and memory, making it ideal for operating systems, video games, and critical software. 

C++ is a programming language designed in 1979 by Bjarne Stroustrup.

### 1. Create the code

for example: create the file **program.cpp** with the next lines:
```
#include<iostream>
using namespace std;

int main()
{ 
  cout<<"Hello World, but in c++!\n"; 
  cout<<"This is a simple C++ program that prints a message to the console.\n";
  printf("This is another way to print a message using printf function from C library.\n");

  return 0;
}
```

#includetells the compiler to search for and interpret all the elements


###  2. Compile the code

In the console: **g++ program.cpp  -o  program**

### 3.  Run the Executable

In the console: **./program**
