```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sc
```

Hacer el analisis estadístico de las variables: Millas por galón (mpg) vs potencia (horsepower) in the dataset: millas_por_galón.csv. Contruya el fit de al 95 C.L.

```python
url = 'https://raw.githubusercontent.com/anferivera/Fisica_Computacional_1/refs/heads/main/Sesiones/seaborn/millas_por_galon.csv'
df = pd.read_csv(url)
df
```

# **Analisis estadistico: mpg vs caballos de fuerza**

```python
#Grafico de mpg vs horsepower teniendo en cuenta la cantidad de cinlindros del motor y el origen del auto

sns.set_theme("paper")
sns.scatterplot(data = df, x = "mpg", y = "horsepower" )
sns.regplot(data = df, x = "mpg", y = "horsepower", logx = True, ci = 95,
           color="c", line_kws=dict(color="r"))
```

¿Existe una correlacion entre ambas variables?

```python
co = df["mpg"].corr(df["horsepower"])
print(f"la correlacion entre las variables mpg y horsepower es: {co:.3f}. Lo cual significa que las dos variables tienen una correlacion alta")
```
